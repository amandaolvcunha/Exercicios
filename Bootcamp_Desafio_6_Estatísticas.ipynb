21 -  Obtenha proporção do bilheteria_bruta para os três generos acima.

```python
generos_selecionados= ['Action', 'Comedy','Drama']

df_generos_3 = df_disney[df_disney['genero'].isin(generos_selecionados)]

df_generos_3 = df_generos_3.groupby('genero')[['bilheteria_bruta']].sum()
df_generos_3 = pd.DataFrame(df_generos_3)
df_generos_3['%'] = df_generos_3['bilheteria_bruta'] / df_generos_3['bilheteria_bruta'].sum() *100

print('Proporção entre eles: {}'.format(df_generos_3))

prop_geral = [action, comedy, drama] / df_disney['bilheteria_bruta'].sum()

print('Proporção geral: {}'.format(prop_geral))


```
